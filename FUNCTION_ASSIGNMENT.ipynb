## 3. What are the different ways to define and call a function in Python?

ANSWER->

**Function Definition in Python**

In Python, functions can be defined in the following main ways:

1. Standard Function Definition
This involves using the def keyword.

A. It creates a named block of reusable code.

B. The function may take parameters and may return a value.

C. It is the most common way to define a function.

2. Anonymous Function (Lambda Function)
Defined using the lambda keyword.

A. It does not have a name unless explicitly assigned.

B. It is generally used for simple, one-line operations.

C. It supports only a single expression and does not use the return statement.

**Function Calling in Python**


Functions in Python can be invoked or "called" in several ways, depending on how they are defined and what kind of arguments they accept:

1. Function Call Without Arguments
A function that does not require any input can be called directly with empty parentheses.

2. Function Call With Positional Arguments

* Arguments are passed in the same order as parameters are defined in the function.

* The position of the argument determines which parameter it corresponds to.

3. Function Call With Keyword Arguments

* Arguments are passed using the names of the parameters.

* This provides clarity and allows arguments to be passed in any order.

4. Function Call Using Default Values

* If the function has parameters with default values, those parameters can be omitted during the function call.

* The function will use the default values provided in the definition.

5. Function Call With Variable-Length Arguments

* Some functions are designed to accept a variable number of arguments.

* Positional variable-length arguments are collected into a tuple.

* Keyword variable-length arguments are collected into a dictionary.

* This allows the function to be more flexible in handling input.

6. Calling Anonymous (Lambda) Functions

* These are called in place, often immediately after definition, or via a variable if assigned.

* They are commonly used where a simple function is needed temporarily.


## 4. What is the purpose of the `return` statement in a Python function?
   ANSWER->

  * **Purpose of the return Statement**

The return statement is used to:

1. Send a Result Back to the Caller
* It allows a function to pass data or a value back to the part of the program where the function was called.

This is how a function produces output.

2. Terminate the Function Execution

* When Python encounters a return statement, it immediately exits the function.

* Any code after the return within the function is not executed.

3. Provide Output for Further Use

* The value returned by a function can be:

* Stored in a variable

* Used in expressions

* Passed as input to another function

4. Return Multiple Values

* The return statement can return multiple values as a tuple, allowing functions to send back more than one result.

5. Return Nothing (None)

* If a function does not have a return statement (or uses return with no value), Python automatically returns a special value called None.

* This indicates no result was explicitly provided.

## 6. Explain the concept of generators in Python and how they are defined.

ANSWER->

What is a Generator?

* A generator is a special type of function that produces values one at a time, instead of returning all results at once.

* It remembers its state, so it can continue from where it left off.

* It is used for large data or infinite sequences, where storing all values at once is not efficient.

How is a Generator Defined?

* It is defined like a normal function, but uses the yield keyword instead of return.

* Each time the generator is called, it pauses at yield, returns a value, and resumes later from the same point.

## 7. What are the advantages of using generators over regular functions?

ANSWER->

1. Memory Efficiency

Generators use less memory because they yield one value at a time, instead of storing all values in memory.

2. Lazy Evaluation

Values are generated only when needed, making them ideal for large datasets or infinite sequences.

3. Improved Performance

Generators improve performance by avoiding the overhead of creating and storing large collections of data.

4. Simplified Code

Generators simplify iteration and state management with the yield keyword, making the code cleaner and more readable.

5. Infinite Sequences

Generators can handle infinite sequences or data streams without running out of memory.

##  1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

ANSWER->




```python
def sum_of_evens(number):
  total = 0
  for num in number:
    if num % 2 == 0:
      total += num
  return total
```

## 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

 ANSWER->

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

ANSWER->



```python
class Fibonacci:
    def __init__(self, limit):
        self.limit = limit
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.limit:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration
```

## 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

ANSWER->


```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
```

## 7. Implement a generator function that reads a file line by line and yields each line as a string.

ANSWER->


```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```
